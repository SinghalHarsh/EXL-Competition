```python
hline=1-train[train["Churn"]==1].shape[0]/train.shape[0]
```

```python
def stacked_plot(X, Y="Churn", a=8, b=3, rows=1, columns=2):
    rcParams['figure.figsize'] = (a, b)
    fig, (ax1, ax2) = plt.subplots(rows,columns)
    var = train.groupby([X, Y]).XX.sum()
    b=var.unstack().plot(kind='bar',stacked=True,ax=ax1)
#     print(var.unstack())
    b.set_title(X+" (Bar Chart)",weight="bold").set_fontsize('12')
    var2 = pd.crosstab(train[X], train[Y]).apply(lambda r: r/r.sum(), axis=1)
    k=var2.plot(kind='bar', stacked=True, ax=ax2, )
    k.axhline(hline, color="k")
    k.set_title(X+" (Stacked Chart)",weight="bold").set_fontsize('12')
#     Columns.remove(X)
```

```python
def scatter(X,Y="Churn", a=8 , b=3,alpha=.3):
    rcParams['figure.figsize'] = (a, b)
    plt.scatter(train[X], train["Churn"], alpha=alpha,)
    plt.title(X,weight="bold").set_fontsize('15')
    plt.show()
#     Columns.remove(X)
```
