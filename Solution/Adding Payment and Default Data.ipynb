```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
Payment = pd.read_csv('Payment.csv')
Default = pd.read_csv('Default.csv')
```

```python
data=pd.merge(Payment, Default, on="Account Number", how="outer" )
```

```python
data["Mode"]=data[["Mode_Month_3","Mode_Month_2","Mode_Month_1"]].mode(axis=1)

del data["Mode_Month_1"], data["Mode_Month_2"], data["Mode_Month_3"]
del data["Day_Month_1_x"], data["Day_Month_2_x"], data["Day_Month_3_x"]
del data["Day_Month_1_y"], data["Day_Month_2_y"], data["Day_Month_3_y"]
```

> if Sum_Month is not 0 then turning down Nan Default_sum to 0.

```python
data["Sum_1"]=np.where((data["Sum_Month_1"].isnull()) & (data["Default_sum_Month_1"].notnull()),0,data["Sum_Month_1"])
data["Sum_2"]=np.where((data["Sum_Month_2"].isnull()) & (data["Default_sum_Month_2"].notnull()),0,data["Sum_Month_2"])
data["Sum_3"]=np.where((data["Sum_Month_3"].isnull()) & (data["Default_sum_Month_3"].notnull()),0,data["Sum_Month_3"])
```

```python
data["Default_Sum_1"]=np.where((data["Sum_Month_1"].notnull()) & (data["Default_sum_Month_1"].isnull()),0,data["Default_sum_Month_1"])
data["Default_Sum_2"]=np.where((data["Sum_Month_2"].notnull()) & (data["Default_sum_Month_2"].isnull()),0,data["Default_sum_Month_2"])
data["Default_Sum_3"]=np.where((data["Sum_Month_3"].notnull()) & (data["Default_sum_Month_3"].isnull()),0,data["Default_sum_Month_3"])
```

```python
del data["Default_sum_Month_1"], data["Default_sum_Month_2"], data["Default_sum_Month_3"]
del data["Sum_Month_1"], data["Sum_Month_2"], data["Sum_Month_3"]
```

.

> If Nan in both Payment and default than filling Null with mean of other 2 months in both Payment and default.

```python
data["X"]=data[["Sum_3","Sum_2","Sum_1"]].sum(axis=1)
data["Y"]=data[["Sum_3","Sum_2","Sum_1"]].count(axis=1)
data["Z"]=data["X"]/data["Y"]

data["Sum_1"]=data["Sum_1"].fillna(data["Z"])
data["Sum_2"]=data["Sum_2"].fillna(data["Z"])
data["Sum_3"]=data["Sum_3"].fillna(data["Z"])

data["X"]=data[["Default_Sum_3","Default_Sum_2","Default_Sum_1"]].sum(axis=1)
data["Y"]=data[["Default_Sum_3","Default_Sum_2","Default_Sum_1"]].count(axis=1)
data["Z"]=data["X"]/data["Y"]

data["Default_Sum_1"]=data["Default_Sum_1"].fillna(data["Z"])
data["Default_Sum_2"]=data["Default_Sum_2"].fillna(data["Z"])
data["Default_Sum_3"]=data["Default_Sum_3"].fillna(data["Z"])

del data["X"], data["Y"], data["Z"]
```

> 

```python
data["Defaults"]=0
data["Defaults"][data["Default_Sum_1"]!=0]=data["Defaults"].apply(lambda x: x+2)
data["Defaults"][data["Default_Sum_2"]!=0]=data["Defaults"].apply(lambda x: x+3)
data["Defaults"][data["Default_Sum_3"]!=0]=data["Defaults"].apply(lambda x: x+6)
```

> Feature Engineering

```python
data["Last_Month_Payment"] = data["Sum_3"]
data["Average_of_payment_of_last_two_month"] = (data["Sum_2"]+data["Sum_3"])/2
data["Average_of_payment_of_last_three_month"] = (data["Sum_2"]+data["Sum_3"] + data["Sum_1"])/3
del data["Sum_1"], data["Sum_2"], data["Sum_3"]
```

```python
data["Last_Month_Default"] = data["Default_Sum_3"]
data["Average_of_default_of_last_two_month"] = (data["Default_Sum_2"]+data["Default_Sum_3"])/2
data["Average_of_default_of_last_three_month"] = (data["Default_Sum_2"]+data["Default_Sum_3"] + data["Default_Sum_1"])/3
del data["Default_Sum_1"], data["Default_Sum_2"], data["Default_Sum_3"]
```
