```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
Payment = pd.read_csv('Payment_Default.csv')
Pattern_Use = pd.read_csv('Pattern_use.csv')
Demographic = pd.read_excel('Demographic_Data.xlsx')
Call_Center = pd.read_csv('Call_Center.csv')
```

```python
Payment.head(100)
```

```python
#Combining Data
```

```python
Payment.shape, Demographic.shape, Pattern_Use.shape, Call_Center.shape
```

```python
main1=pd.merge(Demographic, Payment, on="Account Number", how="left" )
```

```python
main2=pd.merge(main1, Call_Center, on="Account Number", how="left" )
main=pd.merge(main2, Pattern_Use, on="Account Number", how="left" )
```

```python
main.shape
```

```python
main["Commence Date"]=Demographic["Commence Date"]
main["Churn Date"]=Demographic["Churn Date"]

main['Churn Date'] = pd.to_datetime(main['Churn Date'], format='%y-%m-%d')

main["Churn"]=1
main["Churn"][main["Churn Date"].isnull()]=0

del main["Churn Date"]
```

```python
main.to_csv("Combined_Data.csv", index = False)
```
